## Sesion 4 - Ejercicio: Construir calcular_prima() Paso a Paso

Vamos a construir la funcion final de forma incremental.
Cada version agrega una capacidad nueva.

**Resuelve cada nivel antes de ver el siguiente.**

```python
# ── NIVEL 1: Version minima ──────────────────────────────────────────────────
# Solo recibe suma y tasa, calcula prima pura
# (sin recargo, sin IVA — lo mas basico posible)

# Tu codigo aqui:
def calcular_prima(suma, tasa):
    return suma * tasa
```

```python
# ── NIVEL 2: Agregar recargo de seguridad ────────────────────────────────────
# Prima pura * (1 + recargo)
# El recargo por defecto es 15%

# Tu codigo aqui:
def calcular_prima(suma, tasa, recargo=0.15):
    return suma * tasa * (1 + recargo)
```

```python
# ── NIVEL 3: Agregar IVA ─────────────────────────────────────────────────────
# Prima comercial * iva
# El IVA por defecto es 1.16
# Agrega el docstring

# Tu codigo aqui:
def calcular_prima(suma, tasa, recargo=0.15, iva=1.16):
    """Calcula prima comercial con IVA."""
    return suma * tasa * (1 + recargo) * iva
```

```python
# ── NIVEL 4: Devolver prima anual Y prima mensual ────────────────────────────
# return prima_anual, prima_mensual
# Llamar la funcion y desempaquetar los dos valores

# Tu codigo aqui:
def calcular_prima(suma, tasa, recargo=0.15, iva=1.16):
    """Devuelve (prima_anual, prima_mensual)."""
    anual = suma * tasa * (1 + recargo) * iva
    return anual, anual / 12
```

```python
# ── NIVEL 5: Aplicar a una cartera completa ─────────────────────────────────
# Usa tu funcion final con la siguiente cartera:

cartera = [
    {'nombre': 'Ana Torres',   'suma': 500_000,   'tasa': 0.018},
    {'nombre': 'Luis Perez',   'suma': 1_000_000, 'tasa': 0.031},
    {'nombre': 'Maria Garcia', 'suma': 300_000,   'tasa': 0.022},
    {'nombre': 'Carlos Ruiz',  'suma': 2_000_000, 'tasa': 0.048},
    {'nombre': 'Rosa Mendez',  'suma': 750_000,   'tasa': 0.025},
]

# Imprime una tabla con: Nombre | Prima anual | Prima mensual
# Calcula el total de primas de la cartera

# Tu codigo aqui:
def calcular_prima(suma, tasa, recargo=0.15, iva=1.16):
    """Devuelve (prima_anual, prima_mensual)."""
    anual = suma * tasa * (1 + recargo) * iva
    return anual, anual / 12

print(f'{"Nombre":<18} {"Prima anual":>12} {"Prima mensual":>14}')
print('-' * 48)
for p in cartera:
    anual, mensual = calcular_prima(p['suma'], p['tasa'])
    print(f"{p['nombre']:<18} ${anual:>11,.2f} ${mensual:>13,.2f}")

total = sum(calcular_prima(p['suma'], p['tasa'])[0] for p in cartera)
print('-' * 48)
print(f'{"TOTAL":<18} ${total:>11,.2f}')
```
