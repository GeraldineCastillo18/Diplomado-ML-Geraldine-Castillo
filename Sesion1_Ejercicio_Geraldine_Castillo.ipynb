### Datos

```python
# ── Cartera de autos — carga estos datos ────────────────────────────────────
import pandas as pd

cartera_autos = {
    "poliza":     ["A-101","A-102","A-103","A-104","A-105","A-106","A-107","A-108","A-109","A-110"],
    "asegurado":  ["Sofía R.","Miguel T.","Laura G.","Roberto M.","Carmen P.",
                   "Andrés F.","Lucía H.","Diego V.","Patricia N.","Héctor S."],
    "edad":       [24, 38, 31, 55, 42, 27, 46, 33, 58, 22],
    "vehiculo":   ["Sedán","SUV","Hatchback","SUV","Sedán",
                   "Pick-up","Sedán","Hatchback","SUV","Sedán"],
    "modelo_year":[2022, 2023, 2020, 2024, 2021, 2023, 2019, 2022, 2024, 2023],
    "prima_anual":[3_200, 5_800, 2_900, 7_200, 4_100,
                   6_500, 3_800, 2_750, 8_100, 2_600],
    "deducible":  [5_000, 8_000, 5_000, 10_000, 5_000,
                   8_000, 5_000, 5_000, 10_000, 5_000],
    "siniestros": [0, 1, 0, 0, 2, 1, 0, 0, 1, 0],
    "costo_siniestros": [0, 18_500, 0, 0, 42_000, 11_200, 0, 0, 28_000, 0],
}

df_autos = pd.DataFrame(cartera_autos)
print(f"Cartera cargada: {len(df_autos)} pólizas")
df_autos
```

```python
# ── Tarea 1: Estadísticas descriptivas ──────────────────────────────────────
# Escribe tu solución aquí
print("----ESTADISTICAS DESCRIPTIVAS----")
df_autos.describe().round(2)
```

```python
# ── Tarea 2: Resumen ejecutivo ───────────────────────────────────────────────
# Escribe tu solución aquí

print(" --------RESUMEN EJECUTIVO--------")
print(f"Total de polizas: {len(df_autos)}")
print(f"Edad promedio: {int(sum(df_autos['edad'])/len(df_autos))}")
print(f"Prima anual total: {sum(df_autos['prima_anual']):,}")
print(f"Prima maxima: {df_autos['prima_anual'].max():,}")
print(f"Prima minima: {df_autos['prima_anual'].min():,}")
print(f"Deducible promedio: {sum(df_autos['deducible']/len(df_autos)):,}")
print(f"Deducible maximo: {df_autos['deducible'].max():,}")
print(f"Deducible minimo: {df_autos['deducible'].min():,}")
print(f"Siniestros totales: {sum(df_autos['siniestros'])}")
print(f"Siniestro maximo: {df_autos['siniestros'].max():,}")
print(f"Siniestro minimo: {df_autos['siniestros'].min():,}")
print(f"Costo total de siniestros: {sum(df_autos['costo_siniestros']):,}")
print(f"Costo primedio por siniestro: {sum(df_autos['costo_siniestros'])/len(df_autos):,}")
```

```python
# ── Tarea 3: Filtrar pólizas con siniestros ─────────────────────────────────
# Escribe tu solución aquí
print("-----POLIZAS CON SINIESTROS-------")
df_autos[df_autos['siniestros']>0]
```

```python
# ── Tarea 4: Columna frecuencia_individual ───────────────────────────────────
# Escribe tu solución aquí

print("----COLUMNA FRECUENCIA INDIVIDUAL----")
df_autos['frecuencia_ind'] = (df_autos['siniestros'] > 0).replace({True: 1, False: 0})
df_autos[['poliza', 'siniestros', 'frecuencia_ind']]
```
