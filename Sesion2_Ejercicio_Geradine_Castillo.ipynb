## 9. Ejercicio Integrador de Cierre

**Contexto:** Recibes los datos de 5 polizas de seguro de vida y necesitas
generar un reporte de seguimiento. Este ejercicio usa **todos los tipos** vistos hoy.

```python
# ── Datos del ejercicio ──────────────────────────────────────────────────────
from datetime import date

polizas = [
    {'id': 'VID-2026-001', 'asegurado': 'sofia ramirez mendoza',
     'fecha_nac': date(1990, 3, 15), 'fecha_emision': date(2026, 2, 1),
     'fecha_vence': date(2027, 1, 31), 'suma_aseg': 500_000,
     'prima_anual': 3_200.0, 'activa': True, 'beneficiario': None},

    {'id': 'VID-2026-002', 'asegurado': '  MIGUEL TORRES LUNA  ',
     'fecha_nac': date(1975, 8, 22), 'fecha_emision': date(2026, 1, 15),
     'fecha_vence': date(2027, 1, 14), 'suma_aseg': 1_500_000,
     'prima_anual': 9_450.0, 'activa': True, 'beneficiario': 'Elena Torres'},

    {'id': 'VID-2026-003', 'asegurado': 'Laura Gonzalez',
     'fecha_nac': date(1955, 12, 5), 'fecha_emision': date(2025, 7, 1),
     'fecha_vence': date(2026, 6, 30), 'suma_aseg': 300_000,
     'prima_anual': 4_100.0, 'activa': False, 'beneficiario': 'Roberto Gonzalez'},

    {'id': 'VID-2026-004', 'asegurado': 'carlos ruiz vega',
     'fecha_nac': date(2001, 5, 18), 'fecha_emision': date(2026, 3, 10),
     'fecha_vence': date(2027, 3, 9), 'suma_aseg': 250_000,
     'prima_anual': 1_850.0, 'activa': True, 'beneficiario': None},

    {'id': 'VID-2026-005', 'asegurado': 'Patricia Nava',
     'fecha_nac': date(1968, 9, 30), 'fecha_emision': date(2026, 4, 1),
     'fecha_vence': date(2027, 3, 31), 'suma_aseg': 800_000,
     'prima_anual': 6_700.0, 'activa': True, 'beneficiario': 'Hector Nava'},
]
print(f'{len(polizas)} polizas cargadas')
```

```python
# ── Tarea 1: Limpiar nombres ─────────────────────────────────────────────────
# Algunos nombres tienen espacios extra o estan en minusculas.
# Normaliza todos a formato Title Case sin espacios extra.

for i in polizas:
    nombre_limpio = i['asegurado'].strip().title()
    i['asegurado'] = nombre_limpio
    print(nombre_limpio)
```

```python
# ── Tarea 2: Calcular edad y dias de vigencia ────────────────────────────────
print(f"{'Asegurado':>21}    Edad    Días de vigencia")

for i in polizas:
    edad = int((date.today() - i['fecha_nac']).days/365)
    dias = (i['fecha_vence']-i['fecha_emision']).days
    i['edad'] = edad
    i['dias_vigencia'] = dias
    print(f"{i['asegurado']:>21}    {edad:>4}    {dias:>16}")
```

```python
# ── Tarea 3: Resumen de cartera ──────────────────────────────────────────────
# for i in polizas:
#     sa_total = sum(i['suma_aseg'])
#     prima_total = sum(i['prima_anual'])
#     if i['activa'] is True
print(" -------RESUMEN DE CARTERA------")

print(f"Polizas totales: {len(polizas)}")
prima_total = 0
sa_total = 0
edad_total = 0
activos = 0
no_activos = 0
benef = 0
no_benef = 0
for i in polizas:
    prima_total += i['prima_anual']
    sa_total += i['suma_aseg']
    edad_total += i['edad']
    if i['activa'] is True:
        activos += 1
    if i['activa'] is False:
        no_activos += 1
    if i['beneficiario'] is not None:
        benef += 1
    if i['beneficiario'] is None:
        no_benef += 1


print(f"Prima total: {prima_total:,}")
print(f"Suma Asegurada Total: {sa_total:,}")
print(f"Edad promedio: {edad_total/len(polizas):,}")
print(f"Activas: {activos}  |  No activas: {no_activos}")
print(f"Con beneficiario: {benef}    |    Sin beneficiario: {no_benef}")

print("                   ")
print(f"---- POLIZAS ACTIVAS: {activos}")
print(f"{'ID':>12}{'Asegurado':>25}")
for i in polizas:
    if i['activa'] is True:
        print(f"{i['id']:>12} {i['asegurado']:>25}")

print("                   ")
print(f"---- POLIZAS NO ACTIVAS: {no_activos}")
print(f"{'ID':>12}{'Asegurado':>25}")
for i in polizas:
    if i['activa'] is False:
        print(f"{i['id']:>12} {i['asegurado']:>25}")

print("                   ")
print(f"---- POLIZAS CON BENEFICIARIO: {benef}")
print(f"{'ID':>12}{'Asegurado':>25}{'Beneficiario':>25}")
for i in polizas:
    if i['beneficiario'] is not None:
        print(f"{i['id']:>12} {i['asegurado']:>25} {i['beneficiario']:>25}")

print("                   ")
print(f"---- POLIZAS SIN BENEFICIARIO: {no_benef}")
print(f"{'ID':>12}{'Asegurado':>25}{'Beneficiario':>25}")
for i in polizas:
    if i['beneficiario'] is None:
        print(f"{i['id']:>12} {i['asegurado']:>25} {i['beneficiario']}")
```

```python
# ── Tarea 4 (Bonus): Avisos de vencimiento ───────────────────────────────────

tiempo = 90

print(f"Polizas proximas a vencer en {tiempo} dias")
print("----------------------------------------------")
print(f"{'ID':>12} {'Asegurado':>20} {'fecha de vencimiento':>25}")

for i in polizas:
    dias_restantes = (i['fecha_vence'] - date.today()).days
    if dias_restantes < tiempo:
        print(f"{i['id']:>12} {i['asegurado']:>20} {i['fecha_vence'].strftime('%d/%m/%Y'):>25}")
```
